```python
import pandas as pd
import numpy as np

import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/fenago/MLEssentials2/main/datasets/WA_Fn-UseC_-Telco-Customer-Churn.csv")
df.head()
```

```python
df.head().T
```

```python
df.shape
```

```python
df.dtypes
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df['TotalCharges'] = df['TotalCharges'].fillna(0)
```

```python
df.dtypes
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')

string_columns = list(df.dtypes[df.dtypes == 'object'].index)

for col in string_columns:
    df[col] = df[col].str.lower().str.replace(' ', '_')
```

```python
df.dtypes
```

```python
df.churn = (df.churn== 'yes').astype(int)
```

```python
df['churn'].head(10)
```

```python
df.dtypes
```

```python
from sklearn.model_selection import train_test_split
df_train_full, df_test = train_test_split(df, test_size=0.2, random_state=1,stratify=df['churn'])
```

```python
df_train, df_val = train_test_split(df_train_full, test_size=0.33, random_state=11,stratify=df_train_full['churn'])
```

```python
y_train = df_train.churn.values
y_val = df_val.churn.values
```

```python
del df_train['churn']
del df_val['churn']
```

```python
from IPython.display import display
global_mean = df_train_full.churn.mean()
global_mean
```

```python
for col in categorical:
  df_group = df_train_full.groupby(by=col).churn.agg(['mean'])
  df_group['diff'] = df_group['mean'] - global_mean
  df_group['risk'] = df_group['mean'] / global_mean
  display(df_group)
```

```python
from sklearn.metrics import mutual_info_score
def calculate_mi(series):
    return mutual_info_score(series, df_train_full.churn)

df_mi = df_train_full[categorical].apply(calculate_mi)
df_mi = df_mi.sort_values(ascending=False).to_frame(name='MI')


display(df_mi.head())
display(df_mi.tail())
```

```python
df_train_full[numerical].corrwith(df_train_full.churn).to_frame('correlation')
```

# Feature Engineering

```python
from sklearn.feature_extraction import DictVectorizer
```

```python
train_dict = df_train[categorical + numerical].to_dict(orient='records')
```

```python
type(train_dict)
```

```python
train_dict[5]
```

```python
dv = DictVectorizer(sparse=False)
dv.fit(train_dict)
```

```python
X_train = dv.transform(train_dict)
```

```python
df_train_full.shape
```

```python
X_train.shape
```

```python
dv.get_feature_names()
```

```python
sum(df_train_full[categorical].nunique())
```

# Finally... build the model

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(solver='liblinear', random_state=1)
model.fit(X_train, y_train)
```

```python
val_dict = df_val[categorical + numerical].to_dict(orient='records')
X_val = dv.transform(val_dict)
```

```python
X_val.shape
```

```python
model.predict_proba(X_val)
```

```python
y_pred = model.predict_proba(X_val)[:, 1]
```

```python
y_pred
```

```python
churn = y_pred > 0.5
```

```python
(y_val == churn).mean()
```
